### Data Cleaning

#### Filling out the Null values

```python
# You can either drop the null values or use Mean/Median to fill them
train["Mileage"].fillna(train["Mileage"].mode()[0], inplace = True)
train["Engine"].fillna(train["Engine"].mode()[0], inplace = True)
train["Power"].fillna(train["Power"].mode()[0], inplace = True)
train["Seats"].fillna(train["Seats"].mode()[0], inplace = True)

test["Engine"].fillna(test["Engine"].mode()[0], inplace = True)
test["Power"].fillna(test["Power"].mode()[0], inplace = True)
test["Seats"].fillna(test["Seats"].mode()[0], inplace = True)
```

#### Removing SI Units from Mileage, Engine, Power

```python
train["Mileage"] = train["Mileage"].str.split(" ").str[0]
train["Mileage"] = train["Mileage"].astype(float)

test["Mileage"] = test["Mileage"].str.split(" ").str[0]
test["Mileage"] = test["Mileage"].astype(float)

train["Engine"] = train["Engine"].str.split(" ").str[0]
train["Engine"] = train["Engine"].astype(int)

test["Engine"] = test["Engine"].str.split(" ").str[0]
test["Engine"] = test["Engine"].astype(int)

train["Power"] = train["Power"].str.split(" ").str[0]
# Replacing the string Null values in the Power Column
train.loc[train["Power"] == "null", "Power"] = train["Power"].mode()[0]
train["Power"] = train["Power"].astype(float)

test["Power"] = test["Power"].str.split(" ").str[0]
# Replacing the string Null values in the Power Column
test.loc[test["Power"] == "null", "Power"] = test["Power"].mode()[0]
test["Power"] = test["Power"].astype(float)
```

### Encoding Features

```python
# Handling Names
train["Name"] = train["Name"].str.split(" ").str[0]
test["Name"] = test["Name"].str.split(" ").str[0]
```

```python
le = LabelEncoder()
# Training Data
train["Name"] = le.fit_transform(train["Name"])
train["Location"] = le.fit_transform(train["Location"])
train["Fuel_Type"] = le.fit_transform(train["Fuel_Type"])
train["Transmission"] = le.fit_transform(train["Transmission"])
train["Owner_Type"] = le.fit_transform(train["Owner_Type"])
# Testing Data
test["Name"] = le.fit_transform(test["Name"])
test["Location"] = le.fit_transform(test["Location"])
test["Fuel_Type"] = le.fit_transform(test["Fuel_Type"])
test["Transmission"] = le.fit_transform(test["Transmission"])
test["Owner_Type"] = le.fit_transform(test["Owner_Type"])
```

## Modelling

```python
target = train["Price"]
input = train.drop(columns= ["Price"])
x_test = test
```

#### XGBoost

```python
model_xgb = xgb.XGBRegressor(colsample_bytree=0.6,
                             learning_rate=0.05,
                             max_depth=5,
                             n_estimators=1100)

score = cross_val_score(model_xgb, input, target, cv=k_fold, n_jobs=1)
print(score, "\n")
print("XGB Score:",round(np.mean(score)*100, 2), "\n")
```

```python
model_xgb.fit(input,target)
y_pred = model_xgb.predict(x_test)
print(y_pred)
```
